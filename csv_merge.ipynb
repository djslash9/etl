```python
import pandas as pd
import glob
import os

# Get all CSV files in a directory
csv_files = glob.glob('G:/Decisionomics/2025/Dashboard/Template/SPROUT/messages/*.csv')

# Method 1: Simple concatenation
combined_df = pd.DataFrame()

for file in csv_files:
    df = pd.read_csv(file)
    combined_df = pd.concat([combined_df, df], ignore_index=True)

print(f"Combined {len(csv_files)} files with {len(combined_df)} rows")
combined_df.head()
```

```python
combined_df.to_csv('messages.csv', index=False)
```
